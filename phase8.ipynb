```python
#  TRANSFER LEARNING WITH MOBILENETV2 (CIFAR-10 Example)
# This script demonstrates feature extraction + fine-tuning
# using a pretrained MobileNetV2 model on the CIFAR-10 dataset.

# IMPORT LIBRARIES
# -----------------------------
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# Check TensorFlow version and GPU
print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))

#  LOAD & PREPARE DATA
# -----------------------------
# CIFAR-10 → 60k color images (32x32x3) with 10 classes
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()

# Normalize pixel values (range 0–255 → 0–1)
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0

# Class names for reference
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Visualize first 10 images
plt.figure(figsize=(10, 3))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.imshow(X_train[i])
    plt.axis("off")
    plt.title(class_names[y_train[i][0]], fontsize=8)
plt.suptitle("Sample CIFAR-10 Images", fontsize=12)
plt.show()

# LOAD PRETRAINED MODEL
# -----------------------------
# Load MobileNetV2 pretrained on ImageNet (excluding top layers)
base_model = keras.applications.MobileNetV2(
    input_shape=(96, 96, 3),      # ≥32x32
    include_top=False,            # remove original classifier
    weights="imagenet"            # use pretrained weights
)

# Freeze base model (no training yet)
base_model.trainable = False

print("\nBase Model Summary (Frozen):")
base_model.summary()

# RESIZE CIFAR IMAGES
# -----------------------------
# MobileNet expects larger images (≥96x96)
resize_layer = keras.layers.Resizing(96, 96)
X_train_resized = resize_layer(X_train) #resiize ur models inputs yo alrerady trained model
X_test_resized = resize_layer(X_test)

# BUILD FULL MODEL (ie add new dense layer on top of old data to train them )
# -----------------------------
model = keras.Sequential([
    resize_layer,                         # Step 1: Resize 32x32 → 96x96
    base_model,                           # Step 2: Use pretrained CNN base
    layers.GlobalAveragePooling2D(),      # Step 3: Convert features to 1D
    layers.Dense(128, activation="relu"), # Step 4: Custom classification head
    layers.Dropout(0.3),                  # Step 5: Prevent overfitting
    layers.Dense(10, activation="softmax")# Step 6: Output layer (10 classes)
])

# Compile model
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

print("\nFull Model Summary (Feature Extraction Phase):")
model.summary()

# TRAIN NEW CLASSIFIER HEAD (only train head now , later go for thge deeper layers)
# -----------------------------
print("\n Training only new Dense layers (Feature Extraction Phase)...\n")

history = model.fit(
    X_train_resized, y_train,
    validation_split=0.1,   # use 10% for validation
    epochs=5,
    batch_size=64,
    verbose=2
)

# EVALUATE INITIAL MODEL
# -----------------------------
test_loss, test_acc = model.evaluate(X_test_resized, y_test, verbose=0)
print(f"\n Test Accuracy (Feature Extraction Phase): {test_acc:.4f}")

# FINE-TUNING PHASE
# -----------------------------
print("\n Starting Fine-Tuning (unfreezing deeper layers)...")

# Unfreeze the base model
base_model.trainable = True

# Freeze all but last 30 layers to avoid overfitting
for layer in base_model.layers[:-30]:
    layer.trainable = False

# Recompile with smaller LR (important!)
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Fine-tune entire model
fine_tune_history = model.fit(
    X_train_resized, y_train,
    validation_split=0.1,
    epochs=3,
    batch_size=64,
    verbose=2
)

#  FINAL EVALUATION
# -----------------------------
fine_tune_loss, fine_tune_acc = model.evaluate(X_test_resized, y_test, verbose=0)
print(f"\n Test Accuracy (After Fine-Tuning): {fine_tune_acc:.4f}")

#  PLOT LEARNING CURVES
# -----------------------------
plt.figure(figsize=(8, 4))
plt.plot(history.history["accuracy"], label="Train Acc (Feature Extraction)")
plt.plot(history.history["val_accuracy"], label="Val Acc (Feature Extraction)")
plt.plot(np.arange(5, 8), fine_tune_history.history["accuracy"], label="Train Acc (Fine-Tuning)")
plt.plot(np.arange(5, 8), fine_tune_history.history["val_accuracy"], label="Val Acc (Fine-Tuning)")
plt.title("Training & Fine-Tuning Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
